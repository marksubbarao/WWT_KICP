```python
import healpy
import numpy
import matplotlib
%matplotlib inline
```

###Loop through and process all the .fits files in the same directory as the nootbook

```python
files  = !ls *.fits
```

```python
for f in files:
    hmap = healpy.fitsfunc.read_map(f)
    m = healpy.cartview(hmap,title = "", return_projected_map=True,notext=True,cbar=False,xsize=6000, margins= None, cmap="RdBu")
    matplotlib.pyplot.imsave(f[:-5]+'.png', m, cmap="RdBu")
```

###Code for porcessing just one map
####Load Map

```python
hmap = healpy.fitsfunc.read_map('map_lmax2500_nside1024_interp0.3_method1_1_sim0_phi_map.fits')
m = healpy.cartview(hmap,title = "", return_projected_map=True,notext=True,cbar=False,xsize=6000, margins= None, cmap="RdBu")
```

####Map histogram

```python
matplotlib.pyplot.hist(numpy.reshape(m,-1),100)
```

####Save Map

```python
 matplotlib.pyplot.imsave('map_lmax2500_nside1024_interp0.3_method1_1_sim0_phi_map_cart.png', m, cmap="RdBu")
    
```
